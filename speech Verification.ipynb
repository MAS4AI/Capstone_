```python
import sounddevice as sd
import numpy as np
import torch
import torchaudio

duration = 5  # Duration in seconds
sample_rate = 16000  # Sample rate in Hz

print("Recording started...")
recording = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1)
sd.wait()  # Wait until recording is finished
print("Recording finished.")

# Convert recording to a mono waveform as Tensor
waveform = torch.from_numpy(np.squeeze(recording))

# Save the recording as a WAV file
output_path = "recorded_audio.wav"
torchaudio.save(output_path, waveform, sample_rate)

print("Recording saved as", output_path)
```

```python
import sounddevice as sd
import soundfile as sf
def recored_audio(d=10):
    # Set the recording parameters
    duration = d  # Duration in seconds
    sample_rate = 16000  # Sample rate in Hz

    # Record audio
    print("Recording started...")
    recording = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1)
    sd.wait()  # Wait until recording is finished
    print("Recording finished.")

    # Save the recording as a WAV file
    output_path = "recorded_audio.wav"
    sf.write(output_path, recording, sample_rate)

    print("Recording saved as", output_path)
```

```python
from pydub import AudioSegment
import matplotlib.pyplot as plt

# Load the audio file
audio = AudioSegment.from_file("omar1.wav")

# Extract the audio signal
signal = audio.get_array_of_samples()

# Get the sample rate
sample_rate = audio.frame_rate

# Get the duration in seconds
duration = len(signal) / sample_rate

# Create a time axis
time = [t / sample_rate for t in range(len(signal))]

# Plot the audio signal
plt.figure(figsize=(10, 4))
plt.plot(time, signal)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Audio Signal")
plt.show()
```

```python
from pyAudioAnalysis import audioSegmentation
import matplotlib.pyplot as plt

# Perform speaker diarization
segments = audioSegmentation.speaker_diarization("omar1.wav",n_speakers=1)

# Visualize the segments
plt.figure(figsize=(10, 4))
for segment in segments:
    start, end = segment[0], segment[1]
    plt.axvline(x=start, color='red', linestyle='--')
    plt.axvline(x=end, color='green', linestyle='--')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Speaker Diarization")
plt.show()
```
